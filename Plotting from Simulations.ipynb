```python
path = "single trust"

# initialize an empty dictionary to store file contents
file_dict2 = {}

# iterate over all files in the directory
for file_path in glob.glob(os.path.join(path, "*")):
    filename = os.path.basename(file_path)
    with open(file_path, "r") as f:
        file_contents = f.readlines()
     
        
        ttime, ttrust = [], []
        
        for line in file_contents:
            larr = line.split(' ')
            trust_score = int(larr[6])
            _time = int(larr[3])
            if trust_score > 0:
                ttime.append(_time)
                ttrust.append(trust_score)
        
        user_count = int(file_contents[0].split(' ')[2])
    # add the file contents to the dictionary with filename as key
    file_dict2[user_count] = list(zip(ttime, ttrust))
    



#binning by trust scores is actually the way
def get_grouped2(index, file_dict=file_dict2):
    grouped = [[],[],[],[],[]]
    for i in file_dict[index]:
        j,k = i[0], i[1]
        
        if k <= 200:
#             if j < 100:
            grouped[0].append(j)
        elif k <= 400:
            grouped[1].append(j)
        elif k <= 600:
            grouped[2].append(j)
        elif k <= 800:
            grouped[3].append(j)
        else:
            grouped[4].append(j)
            
            
        #get avg_times
    avg_grouped = [median(i) for i in grouped]
    [print(f"trust score {200 + 200*j}, count:{index} with median {median(i)} and std_dev: {statistics.stdev(i)}") for j,i  in enumerate(grouped)]
    return avg_grouped




species = ['0-200', '201-400', '401-600', '601-800', '801-1000'] #make this task count


# create the first bar plot

weight_counts = {
    '10': get_grouped2(10),
    '30': get_grouped2(30),
    '50': get_grouped2(50),
    '100': get_grouped2(100),
    '200':get_grouped2(200)
}


width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(5)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_xlabel('Trust Score')
ax.set_ylabel('Response Time(s)')
ax.set_title('Single Server')    

ax.set_ylim(0, 350)

ax.legend(loc="upper left", ncols=3)

plt.show()
```
